### Modelos determinísticos 

La salida del modelo está determinada por los valores de los parámetros y la condiciones iniciales. Los modelos deterministas tienen la ventaja de ser susceptibles de análisis matemáticos, como en el caso del uso de ecuaciones diferenciales ordinarias para modelar la propagación geográfica de enfermedades infecciosas dentro de las poblaciones humanas. La simulación de estos modelos a través de parámetros y ajustes de entrada puede proporcionar información importante sobre las características cuantitativas y cualitativas de las epidemias (Sattenspiel, 2009).[1]

### Modelos estocásticos

En constraste, los modelos estocásticos, poseen cierta aleatoriedad inherente. El mismo conjunto de valores de parámetros y condiciones iniciales conducirán a un conjunto de diferentes salidas (ej. stock market).

### Repaso de algunos conceptos

**Probability space:** Matemáticamente, el espacio de probabilidad tiene tres componentes ($\Omega, F, P$). [Ref](http://www.its.caltech.edu/~mshum/stats/lect1.pdf).

$\Omega$: Espacio muestral. Conjunto de salidas de un experimento. 
* Ejemplo: Lanzamiento de una moneda, dos veces.

$\Omega = \{HH, HT,TH, TT\}$

**$F$: ** El espacio de eventos contiene los subconjuntos de eventos posibles, de $\Omega$.

* Ejemplo: "Al menos una $H$" el resultado sería $\{HH, HT, TH\}$.

**$P$:** La función de probabilidad asigna un número ("probabilidad") a un evento en $B$.

**Random variables:** $X: \Omega \rightarrow \Re $ es una función que asigna valores reales a las salidad de $\Omega$

**Random process or Stochastic process:**: es una familia de variables aleatorias $X(t): t \in T$, donde $t$  usualmente denota tiempo.
